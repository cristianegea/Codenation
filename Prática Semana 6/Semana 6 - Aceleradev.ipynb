## Analysis

### Seleção por completude (sem _missing values_)

```python
# Criação de um dataframe auxliar para analisar a consistência das variáveis
cons = pd.DataFrame({'colunas' : df.columns,
                    'tipo': df.dtypes,
                    'missing' : df.isna().sum(),
                    'size' : df.shape[0],
                    'unicos': df.nunique()})

# % de missing values
cons['percentual'] = round(cons['missing'] / cons['size'],2)
```

```python
cons.head()
```

```python
# Histograma com frequência de missing values => a maior parte dos dados estará entre 0 e 20%
cons.percentual.plot.hist( bins = 5)
```

### Removendo colunas com dados missing

```python
print('Contagem de colunas com ATÉ 20% de dados faltantes', cons[cons.percentual < 0.2].shape[0])
print('Contagem de colunas com 0% de dados faltantes',  cons[cons.percentual == 0].shape[0])

# de 81 colunas do dataframe, 76 colunas tem até 20% de missing values e 63 colunas sem missing values
```

```python
# Análise das colunas sem missing values
cons[cons.percentual == 0]['tipo'].value_counts()

# 35 colunas contêm inteiros e 28 contêm objetos
```

```python
# Criação de uma coluna informaando se a variável está completa (0% de NA) ou não
cons['completa'] = ['completa' if x == 0 else 'faltante' for x in cons['percentual']]
```

```python
# Manutenção das colunas sem missing values
mantem = list(cons[cons['completa'] == 'completa']['colunas'])
df = df[mantem]
```

```python
# Criação de uma lista de colunas numéricas
colunas_numericas = list(cons[((cons['tipo'] != 'object') &
                              (cons['completa'] == 'completa'))]['colunas'])
```

Um ponto importante em uma análise de dados é chegar a um modelo considerado _baseline_ (para ser utilizado como benchmark para novos experimentos).

## Analisando as features com yellowbrick

```python
pip install yellowbrick

# yellowbrick => biblioteca focada em machine learning
```

```python
# Remoção da coluna ID das colunas numéricas (para não ter o risco de interferir nas estimações de treino e de teste)
colunas_numericas.remove('Id')
df = df[colunas_numericas]
```

```python
# "SalePrice" => coluna que eu quero prever
y_train = df['SalePrice']                       # dataframe "df" com "SalePrice"
X_train = df.drop(columns = 'SalePrice')        # dataframe "df" sem "SalePrice"

# o mesmo tratamento direcionado aos dados de treino são direcionados aos dados de teste
# O tratamento dos dados de teste é realizado em cima do treinamento realizado nos dados de treino
```

```python
# Rank de features com 1 dimensão (mostra quais features são mais importantes)
from yellowbrick.features import Rank1D

visualizer = Rank1D(algorithm='shapiro')

visualizer.fit(X_train, y_train)           
visualizer.transform(X_train)        
visualizer.show()              
```

```python
# Análise de componentes principais (com as variáveis com têm correlação > 0.6 em relação a "SalePrice")
from yellowbrick.features import PCA


visualizer = PCA(scale=True, proj_features=True, projection=2)      # projection => nº de dimensões (nº de PCAs)
visualizer.fit_transform(X_train[correlacionadas], y_train)
visualizer.show()
```

```python
# Correlação entre features (correlação entre a variável de treino) => dá uma noção de dimensionalidade entre as features
from yellowbrick.target import FeatureCorrelation

features = list(X_train.columns)

visualizer = FeatureCorrelation(labels=features)

visualizer.fit(X_train, y_train)        
visualizer.show()       
```

## Treinando o modelo

```python
pip install sklearn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()

# LinearRegression => OLS
```

```python
# Ajustamento do modelo de regressão
reg.fit(X_train, y_train)
```

```python
colunas_treinamento = X_train.columns
```

```python
# leitura dos dados de teste
X_test = pd.read_csv('test.csv')
y_test = pd.read_csv('sample_submission.csv')
```

```python
y_test = y_test['SalePrice']
```

```python
# Substituição dos missing values por média (se houver NA)
X_test= X_test[colunas_treinamento].fillna(df[colunas_treinamento].mean())
```

```python
# Predições
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
erro_normal = mean_squared_error(y_pred=y_pred, y_true=y_test)
```

```python
erro_normal
```

```python
# Gráfico de dispersão entre o y projetado e o Y de teste
sns.scatterplot(y_pred, y_test)

# se os dados forem altamente correlacionados, então os dados de teste coincidirão com os dados previtos
```

## Aplicando o Feature Selection

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(reg)

# RFE => elimina recurssivamente as features
```

```python
rfe.fit(X_train, y_train)
```

```python
# Dataframe indicando quais se a feature foi escolhida ou não para o modelo
pd.DataFrame({'coluna':X_train.columns,
              'bool': rfe.get_support(),
              'coeficientes': pd.Series(reg.coef_)})
```

```python
# Transformação do aprendizado adquirido no treino em dados => transformação das features feitas com dados de treino em dados de teste
X_train_importante = rfe.transform(X_train)
X_test_importante = rfe.transform(X_test)

# importante: a feature não pode ser feita com dados de teste, mas deve ser feita com dados de treino
```

```python
# Ajustamento do modelo de regressão
reg.fit(X_train_importante, y_train)
```

```python
# predição
y_pred_imp = reg.predict(X_test_importante)
```

```python
erro_imp = mean_squared_error(y_pred=y_pred_imp, y_true=y_test)
```

```python
erro_imp
```

```python
from yellowbrick.model_selection import RFECV


# Instantiate RFECV visualizer with a linear SVM classifier
visualizer = RFECV(reg)

visualizer.fit(X_train, y_train)        # Fit the data to the visualizer
visualizer.show()           # Finalize and render the figure
```

## Aplicando PCA

```python
pca.explained_variance_ratio_
```

```python
y_pred_pca = reg.predict(X_test_pca)
```
